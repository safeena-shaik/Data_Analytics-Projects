```python
import pandas as pd

df = pd.read_excel("/content/student_performance_data (1).xlsx")
print(df.head())
#average calculation
avg_marks = df["Marks"].mean()
avg_attendance = df["Attendance (%)"].mean()
avg_logins = df["Logins"].mean()

print("Average Marks:", avg_marks)
print("Average Attendance:", avg_attendance)
print("Average Logins:", avg_logins)
#correlation Analysis
correlation_matrix = df[["Marks", "Attendance (%)", "Logins"]].corr()
print(correlation_matrix)
#absentee impact
df["Attendance_Level"] = df["Attendance (%)"].apply(lambda x: "Low" if x < 75 else "High")
impact = df.groupby("Attendance_Level")["Marks"].mean()
print(impact)
#visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Bar chart: Top vs Struggling Students
top_students = df.sort_values("Marks", ascending=False).head(5)
struggling_students = df.sort_values("Marks").head(5)

plt.figure(figsize=(10, 5))
sns.barplot(data=top_students, x="Student_ID", y="Marks", color="green", label="Top")
sns.barplot(data=struggling_students, x="Student_ID", y="Marks", color="red", label="Struggling")
plt.title("Top vs Struggling Students")
plt.legend()
plt.show()

# Heatmap of correlations
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
