```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
df = pd.read_excel("/content/customer_segmentation_data.xlsx")

# Drop Customer_ID for clustering
X = df.drop("Customer_ID", axis=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
from sklearn.cluster import KMeans

# Use 4 clusters as an example
kmeans = KMeans(n_clusters=4, random_state=42)
df["Cluster"] = kmeans.fit_predict(X_scaled)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Annual_Income_k", y="Spending_Score", hue="Cluster", palette="Set2", s=100)
plt.title("Customer Segmentation Based on Income & Spending")
plt.xlabel("Annual Income (k)")
plt.ylabel("Spending Score")
plt.legend(title="Cluster")
plt.show()
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df["Age"], df["Annual_Income_k"], df["Spending_Score"], c=df["Cluster"], cmap="Set2", s=60)
ax.set_xlabel("Age")
ax.set_ylabel("Income")
ax.set_zlabel("Spending Score")
plt.title("3D Customer Segmentation View")
plt.show()
```
