```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_excel("/content/sample_covid19_data.xlsx")
#df["Date"] = pd.to_datetime(df["Date"])
print(df.head())
#line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Date", y="Confirmed", hue="Country", marker="o")
plt.title("Cumulative Confirmed COVID-19 Cases")
plt.ylabel("Confirmed Cases")
plt.show()

plt.figure(figsize=(12, 6))
for country in df["Country"].unique():
    sub_df = df[df["Country"] == country]
    plt.fill_between(sub_df["Date"], sub_df["Recovered"], label=country, alpha=0.5)

plt.title("Cumulative Recoveries Over Time")
plt.ylabel("Recovered Cases")
plt.legend()
plt.show()
heatmap_data = df.pivot(index="Date", columns="Country", values="Deaths")

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="Reds")
plt.title("Heatmap of COVID-19 Deaths Over Time")
plt.show()
```
